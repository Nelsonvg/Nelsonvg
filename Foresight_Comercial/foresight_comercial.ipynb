<H1>MODELO DE PREDICCIÓN POR CATEGORIA</H1>

```python
# ==========================================================
# COMBINACIÓN DE MODELOS ÓPTIMOS POR CATEGORÍA (CON SARIMA AJUSTADO)
# ==========================================================

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings("ignore")

# ----------------------------------------------------------
# Función de chequeo de NaN
# ----------------------------------------------------------
def check_nan(label, arr):
    """Valida si hay NaN en un array/serie y lo reporta."""
    if isinstance(arr, (pd.Series, pd.DataFrame)):
        if arr.isna().any().any():
            print(f"⚠️ NaN detectado en {label}")
            return True
    else:
        if np.isnan(arr).any() if hasattr(arr, "__iter__") else np.isnan(arr):
            print(f"⚠️ NaN detectado en {label}: {arr}")
            return True
    return False

# ----------------------------------------------------------
# 1️⃣ Cargar los datos históricos
# ----------------------------------------------------------
archivo_historico = "ventas_mensuales_categorizadas.xlsx"
pivot = pd.read_excel(archivo_historico)
pivot["Mes"] = pd.to_datetime(pivot["Mes"], errors="coerce")
pivot = pivot.sort_values("Mes").set_index("Mes")

check_nan("pivot inicial", pivot)

# ----------------------------------------------------------
# 2️⃣ Identificar automáticamente último mes histórico y mes de predicción
# ----------------------------------------------------------
ultimo_mes_hist = pivot.index[-1]              # último mes en histórico
mes_prediccion = ultimo_mes_hist + pd.offsets.MonthBegin(1)  # mes siguiente

print(f"Último mes histórico: {ultimo_mes_hist.strftime('%Y-%m')}")
print(f"Mes de predicción: {mes_prediccion.strftime('%Y-%m')}")

# ----------------------------------------------------------
# 3️⃣ Cargar los datos reales del mes predicho
# ----------------------------------------------------------
# Se asume que 'datos_reales' contiene columna 'Mes' y 'Categoría'
datos_reales = ventas_reales_mensuales.copy()
check_nan("datos_reales", datos_reales)



# ----------------------------------------------------------
# 3️⃣ Definir modelo óptimo por categoría
# ----------------------------------------------------------
modelos_categoria = {
    "Lavadoras": "SARIMA",
    "Refrigeración Comercial": "Lineal",
    "Refrigeración Doméstica": "Lineal",
    "Televisión": "RandomForest",
    "Cocción": "Lineal"
}

# ----------------------------------------------------------
# 4️⃣ Parámetros óptimos SARIMA por categoría
# ----------------------------------------------------------
parametros_por_categoria = {
    "Lavadoras": {'order': (0,1,1), 'seasonal_order': (0,1,1,12)},
    "Refrigeración Comercial": {'order': (1,1,1), 'seasonal_order': (1,1,1,12)},
    "Refrigeración Doméstica": {'order': (0,1,2), 'seasonal_order': (0,1,1,12)},
    "Televisión": {'order': (1,1,1), 'seasonal_order': (0,1,1,12)},
    "Cocción": {'order': (1,1,0), 'seasonal_order': (1,1,0,12)}
}

# ----------------------------------------------------------
# 5️⃣ Entrenar y predecir por categoría
# ----------------------------------------------------------
resultados = {}

for categoria, modelo_nombre in modelos_categoria.items():
    serie = pivot[categoria].dropna().copy()
    check_nan(f"serie histórica {categoria}", serie)

    y_real = datos_reales.loc[
        datos_reales["Categoría"] == categoria, "Cantidad inv."
    ].sum()

    check_nan(f"y_real {categoria}", y_real)

    pred = np.nan  # inicialización

    # ===============================
    # MODELO SARIMA
    # ===============================
    if modelo_nombre == "SARIMA":
        try:
            params = parametros_por_categoria.get(
                categoria, {"order": (1,1,1), "seasonal_order": (0,1,1,12)}
            )
            modelo = SARIMAX(
                serie,
                order=params["order"],
                seasonal_order=params["seasonal_order"],
                enforce_stationarity=False,
                enforce_invertibility=False
            )
            fit = modelo.fit(disp=False)

            start = mes_prediccion
            end = mes_prediccion
            pred = fit.predict(start=start, end=end).iloc[-1]



        except Exception as e:
            print(f"⚠️ Error SARIMA en {categoria}: {e}")
            pred = np.nan

    # ===============================
    # MODELO LINEAL
    # ===============================
    elif modelo_nombre == "Lineal":
        t = np.arange(len(serie)).reshape(-1, 1)
        modelo = LinearRegression()
        modelo.fit(t, serie.values)
        pred = modelo.predict([[len(serie)]])[0]

    # ===============================
    # MODELO RANDOM FOREST
    # ===============================
    elif modelo_nombre == "RandomForest":
        hist = pd.DataFrame({"Mes": serie.index, "Unidades": serie.values}).copy()
        hist["Year"] = hist["Mes"].dt.year
        hist["Month"] = hist["Mes"].dt.month
        hist["Period"] = hist["Mes"].dt.to_period("M")

        def make_lags(df, lags=[1, 3, 12]):
            for lag in lags:
                df[f"lag_{lag}"] = df["Unidades"].shift(lag)
            df["rolling_3"] = df["Unidades"].shift(1).rolling(3).mean()
            return df

        hist = make_lags(hist)
        hist = hist.dropna(subset=["lag_1", "lag_3", "lag_12", "rolling_3"])
        check_nan(f"hist {categoria}", hist)

        target_period = pd.Period("2025-11", freq="M")
        train_df = hist[hist["Period"] < target_period].copy()

        feature_cols = ["Year", "Month", "lag_1", "lag_3", "lag_12", "rolling_3"]
        X = train_df[feature_cols]
        y = train_df["Unidades"]

        model = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
        model.fit(X, y)

        last = hist.iloc[-1]
        lag1 = last["Unidades"]
        lag3 = hist["Unidades"].iloc[-3:].mean()
        lag12 = hist["Unidades"].iloc[-12:].mean() if len(hist) >= 12 else hist["Unidades"].mean()
        rolling_3 = hist["Unidades"].iloc[-3:].mean()

        pred_df = pd.DataFrame([{
            "Year": target_period.year,
            "Month": target_period.month,
            "lag_1": lag1,
            "lag_3": lag3,
            "lag_12": lag12,
            "rolling_3": rolling_3
        }])
        check_nan(f"pred_df {categoria}", pred_df)

        pred = model.predict(pred_df)[0]

    check_nan(f"predicción {categoria}", pred)

    # ===============================
    # Evaluación (solo si no hay NaN)
    # ===============================
    if not (np.isnan(pred) or np.isnan(y_real)):
        mae = mean_absolute_error([y_real], [pred])
        rmse = mean_squared_error([y_real], [pred]) ** 0.5
        mape = np.mean(np.abs((y_real - pred) / y_real)) * 100 if y_real != 0 else 0
    else:
        mae = rmse = mape = np.nan
        print(f"⚠️ Métricas no calculadas en {categoria} por NaN")

    resultados[categoria] = {
        "Categoría": categoria,
        "Modelo": modelo_nombre,
        "Predicción Mes 2025": pred,
        "Real Mes 2025": y_real,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE (%)": mape
    }

# ----------------------------------------------------------
# 6️⃣ Resultados finales
# ----------------------------------------------------------
df_resultados = pd.DataFrame(resultados).T
print("\n===== RESULTADOS COMBINADOS POR MODELO ÓPTIMO (SARIMA ajustado) =====")
print(df_resultados)
```

```python
# ============================================================
# 📊 PREDICCIÓN MES 2025 POR CATEGORÍA - TOP MÁS Y MENOS
# ============================================================

import pandas as pd

# === 1️⃣ Cargar archivos ===
datos_octubre = pd.read_excel("real_mes_categorizado.xlsx")  # Ventas reales

# === 2️⃣ Normalizar nombres de columnas ===
df_resultados.columns = [c.strip().lower().replace(" ", "_") for c in df_resultados.columns]
datos_octubre.columns = [c.strip().lower().replace(" ", "_") for c in datos_octubre.columns]

# Verificación rápida
print("📄 Columnas resultados:", df_resultados.columns.tolist())
print("📄 Columnas ventas:", datos_octubre.columns.tolist())

# === 3️⃣ Validar columnas mínimas ===
if not {"categoría", "cantidad_inv."}.issubset(datos_octubre.columns):
    raise ValueError("⚠️ El archivo 'real_mes_categorizado.xlsx' debe tener columnas: 'categoría' y 'cantidad_inv.'.")

if not {"categoría", "predicción_mes_2025"}.issubset(df_resultados.columns):
    raise ValueError("⚠️ El archivo de predicciones debe tener columnas: 'categoría' y 'predicción_mes_2025'.")

# Normalizar texto en categoría
df_resultados["categoría"] = df_resultados["categoría"].str.strip().str.lower()
datos_octubre["categoría"] = datos_octubre["categoría"].str.strip().str.lower()

# === 4️⃣ Resumen comparativo venta real vs predicción ===
comparativo = datos_octubre.groupby("categoría", as_index=False)["cantidad_inv."].sum()

comparativo = comparativo.merge(
    df_resultados[["categoría", "predicción_mes_2025"]],
    on="categoría", how="left"
)

comparativo["diferencia"] = comparativo["predicción_mes_2025"] - comparativo["cantidad_inv."]
comparativo["error_abs"] = abs(comparativo["diferencia"])
comparativo["%_precisión"] = (1 - comparativo["error_abs"] / comparativo["cantidad_inv."]) * 100
comparativo["%_precisión"] = comparativo["%_precisión"].clip(lower=0)  # nunca negativo

comparativo = comparativo[comparativo["categoría"] != "otro"]

# Reordenar columnas
comparativo = comparativo[[
    "categoría", "predicción_mes_2025", "cantidad_inv.", "%_precisión", "diferencia"
]]

# Redondear antes del rename
cols_redondear = ["predicción_mes_2025", "cantidad_inv.", "%_precisión", "diferencia"]
for col in cols_redondear:
    comparativo[col] = pd.to_numeric(comparativo[col], errors="coerce").round(2)

# Luego renombrar
comparativo.rename(columns={
    "categoría": "Categoría",
    "predicción_mes_2025": "Predicción Mes 2025",
    "cantidad_inv.": "Ventas Reales Mes 2025",
    "%_precisión": "% Precisión",
    "diferencia": "Diferencia"
}, inplace=True)

print("\n✅ Comparativo Venta Real VS Predicción Mes 2025:")
print(comparativo.to_string(index=False))


# === 5️⃣ Función: generar tablas por categoría ===
def generar_tabla_categoria(cat, pred):
    df_cat = datos_octubre[datos_octubre["categoría"] == cat.lower()]
    if df_cat.empty:
        return None, None

    # 🔹 Agrupar por item y sumar unidades
    df_grouped = df_cat.groupby("item", as_index=False)["cantidad_inv."].sum()

    # Ordenar por unidades
    df_sorted = df_grouped.sort_values("cantidad_inv.", ascending=False)

    # Top 5 y Bottom 3
    top5 = df_sorted.head(5)
    bottom3 = df_sorted.tail(3)

    tabla_top = pd.DataFrame({
        "CATEGORÍA": [cat.upper()],
        "CANTIDAD PREDICHA": [round(pred, 2)],
        "TOP 5 MÁS VENDIDOS": [", ".join(top5["item"].astype(str))]
    })

    tabla_bottom = pd.DataFrame({
        "CATEGORÍA": [cat.upper()],
        "CANTIDAD PREDICHA": [round(pred, 2)],
        "TOP 3 MENOS VENDIDOS": [", ".join(bottom3["item"].astype(str))]
    })

    return tabla_top, tabla_bottom


# === 6️⃣ Generar todas las tablas ===
tablas_top, tablas_bottom = [], []

for _, row in df_resultados.iterrows():
    cat = row["categoría"]
    pred = row.get("predicción_mes_2025", None)

    # 🚫 Saltar categoría 'otro'
    if cat == "otro":
        continue
        
    top_df, bottom_df = generar_tabla_categoria(cat, pred)

    if top_df is not None:
        tablas_top.append(top_df)
    if bottom_df is not None:
        tablas_bottom.append(bottom_df)

# Consolidar resultados
top_final = pd.concat(tablas_top, ignore_index=True) if tablas_top else pd.DataFrame()
bottom_final = pd.concat(tablas_bottom, ignore_index=True) if tablas_bottom else pd.DataFrame()

# === 7️⃣ Guardar todo en Excel ===
with pd.ExcelWriter("Informe_gerencial_2025.xlsx") as writer:
    comparativo.to_excel(writer, sheet_name="Comparativo_Ventas", index=False)
    top_final.to_excel(writer, sheet_name="Top_5_Mas_Vendidos", index=False)
    bottom_final.to_excel(writer, sheet_name="Top_3_Menos_Vendidos", index=False)

print("\n📁 Archivo generado: Informe_gerencial_2025.xlsx")
print("Incluye:")
print(" - Comparativo Real vs Predicción por Categoría")
print(" - Top 5 más vendidos por categoría")
print(" - Top 3 menos vendidos por categoría")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose

# 1. Cargar datos
df = pd.read_excel("data_158_categorizado_1.xlsx")  # Ajusta el nombre si es necesario
df['Fecha'] = pd.to_datetime(df['Fecha'])
df = df.set_index('Fecha')

# 2. Función para probar estacionariedad
def test_stationarity(series, name="Serie"):
    result = adfuller(series.dropna())
    print(f"--- {name} ---")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.6f}")
    if result[1] <= 0.05:
        print("✅ La serie es estacionaria")
    else:
        print("❌ La serie NO es estacionaria")

# 3. Analizar por categoría
for cat in df['Categoría'].unique():
    df_cat = df[df['Categoría'] == cat]['Cantidad inv.']

    # Frecuencias
    monthly = df_cat.resample('ME').sum()  # Usa 'ME' en lugar de 'M'
    weekly = df_cat.resample('W').sum()
    daily = df_cat.resample('D').sum()

    # Prueba de estacionariedad
    test_stationarity(monthly, f"{cat} - Mensual")
    test_stationarity(weekly, f"{cat} - Semanal")
    test_stationarity(daily, f"{cat} - Diario")

    # 4. Visualización de series
    fig, axes = plt.subplots(3, 1, figsize=(20, 16))
    monthly.plot(ax=axes[0], title=f"{cat} - Serie Mensual")
    weekly.plot(ax=axes[1], title=f"{cat} - Serie Semanal")
    daily.plot(ax=axes[2], title=f"{cat} - Serie Diaria")
    plt.tight_layout()
    plt.show()

    # 5. Descomposición estacional (mensual)
    decomp = seasonal_decompose(monthly, model='additive', period=12)

    fig = decomp.plot()
    fig.set_size_inches(16, 10)

    plt.suptitle(f"Descomposición estacional: {cat} (Mensual)", fontsize=20, fontweight='bold')

    for ax in fig.axes:
        ax.tick_params(axis='x', labelsize=12, rotation=45)
        ax.tick_params(axis='y', labelsize=12)

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

# Cargar y preparar datos
df = pd.read_excel("data_158_categorizado_1.xlsx")
df['Fecha'] = pd.to_datetime(df['Fecha'])
df = df.set_index('Fecha')

# Crear tabla resumen
calendario = pd.DataFrame()

for cat in df['Categoría'].unique():
    df_cat = df[df['Categoría'] == cat]['Cantidad inv.']
    monthly = df_cat.resample('ME').sum()

    # Descomposición estacional
    decomp = seasonal_decompose(monthly, model='additive', period=12)
    estacional = decomp.seasonal.dropna()

    # Agrupar por mes
    estacional_por_mes = estacional.groupby(estacional.index.month).mean()
    estacional_por_mes.index = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']

    # Normalizar para comparar
    estacional_por_mes = (estacional_por_mes - estacional_por_mes.mean()) / estacional_por_mes.std()

    calendario[cat] = estacional_por_mes

# Transponer para que las filas sean categorías
calendario = calendario.T

# Clasificar meses como "Alto", "Medio", "Bajo"
def clasificar(valor):
    if valor > 0.5:
        return "🔺 Alto"
    elif valor < -0.5:
        return "🔻 Bajo"
    else:
        return "⚪ Medio"

calendario_clasificado = calendario.applymap(clasificar)
print(calendario_clasificado)

# exportar
```
